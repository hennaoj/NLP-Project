```python
import nltk
from nltk.corpus.reader.bnc import BNCCorpusReader
reader = BNCCorpusReader(root='BNC/Texts/', fileids=r'[A-K]/\w*/\w*\.xml')
```

```python
words = reader.tagged_words()
```

```python
love = []
hate = []
like = []
dislike = []

for w in words:
    if w[0].lower() == 'love':
        love.append(w)
    elif w[0].lower() == 'hate':
        hate.append(w)
    elif w[0].lower() == 'like':
        like.append(w)
    elif w[0].lower() == 'dislike':
        dislike.append(w)
```

```python
fdist_love = nltk.FreqDist(w[1] for w in love)
fdist_hate = nltk.FreqDist(w[1] for w in hate)
fdist_like = nltk.FreqDist(w[1] for w in like)
fdist_dislike = nltk.FreqDist(w[1] for w in dislike)
```

```python
tags_love = fdist_love.keys()
tags_hate = fdist_hate.keys()
tags_like = fdist_like.keys()
tags_dislike = fdist_dislike.keys()
```

```python
freqs_love = []
for t in tags_love:
    freqs_love.append(fdist_love.freq(t))
```

```python
freqs_hate = []
for t in tags_hate:
    freqs_hate.append(fdist_hate.freq(t))
```

```python
freqs_like = []
for t in tags_like:
    freqs_like.append(fdist_like.freq(t))
```

```python
freqs_dislike = []
for t in tags_dislike:
    freqs_dislike.append(fdist_dislike.freq(t))
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(20, 20))

axs[0, 0].bar(tags_love, freqs_love, color ='maroon', width = 0.3)
axs[0, 0].set_title('Love (N = %i)' % fdist_love.N())

axs[0, 1].bar(tags_hate, freqs_hate, color ='maroon', width = 0.3)
axs[0, 1].set_title('Hate (N = %i)' % fdist_hate.N())

axs[1, 0].bar(tags_like, freqs_like, color ='maroon', width = 0.3)
axs[1, 0].set_title('Like (N = %i)' % fdist_like.N())

axs[1, 1].bar(tags_dislike, freqs_dislike, color ='maroon', width = 0.3)
axs[1, 1].set_title('Dislike (N = %i)' % fdist_dislike.N())

plt.subplots_adjust(left=0.1, bottom=0.1, right=0.5, top=0.5, wspace=0.2, hspace=0.2)

ylim = (0, 1)

plt.setp(axs, ylim=ylim)

plt.show()
```
